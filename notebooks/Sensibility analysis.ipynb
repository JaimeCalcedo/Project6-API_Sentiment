NLTK con esta librería medimos la polaridad

```python
def sentimentAnalysisPolarity(sentence):
    sia = SentimentIntensityAnalyzer()
    polarity = sia.polarity_scores(sentence)
    return polarity
```

```python
sentimentAnalysisPolarity("I hate you")
```

```python
def sentimentAnalysisCompound(sentence):
    sia = SentimentIntensityAnalyzer()
    polarity = sia.polarity_scores(sentence)
    pol = polarity["compound"]
    return pol
```

```python
todo["sentiment_compund"] = todo["message"].apply(sentimentAnalysisCompound)
todo.head(20)
```

```python
grouped = todo.groupby("user").mean().reset_index()
grouped
```
